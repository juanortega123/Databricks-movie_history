```python
import smtplib
from email.mime.text import MIMEText

subject = 'PipeLine Error'
body = 'El proceso de pipeline ha caido ya que no se ha encontrado la carpeta de salida.'
sender = 'juan.ortega.a@tecsup.edu.pe'
my_password = 'ycmavbwhkpvkdhui'
recipients = [sender]

def send_mail(subject, body, sender, recipients):
    msg = MIMEText(body)
    msg['subject'] = subject
    msg['From'] = sender
    msg['To'] = ', '.join(recipients)

    with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp_server:
        smtp_server.login(sender, my_password)
        smtp_server.sendmail(sender, recipients, msg.as_string())
    print("Message sent!")

send_mail(subject, body, sender, recipients)
```
