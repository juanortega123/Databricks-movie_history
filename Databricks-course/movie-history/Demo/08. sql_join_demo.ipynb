## Inner Join

#### tipos de join
1. inner 
2. Left
3. Rigth
4. Full
5. Semi
6. Anti
7. Cross


####1. Inner

Muestra los dastos que se concatenan en ambas tablas tando en la de movies y la de countries.

```sql
create or replace temp view v_movies_countries
as
select VM.title,VM.budget,VM.revenue,VM.year_release_date,VM.release_date,VC.country_name
from v_movies_2011 VM
inner join v_productions_countries VPC on VM.movie_id = VPC.movie_id
inner join v_countries VC on VPC.country_id = VC.country_id;

select * from v_movies_countries
```

#### 2. Left

muestra la tabla competa de la izquierda (left) sin considerar si hay o no valores que concatenan en la segunda tabla, por ese motivo se ven "null" en la culumna que pertenece a Countries.

```sql
create or replace temp view v_movies_countries_lf
as
select VM.title,VM.budget,VM.revenue,VM.year_release_date,VM.release_date,VC.country_name
from v_movies_2011 VM
left join v_productions_countries VPC on VM.movie_id = VPC.movie_id
left join v_countries VC on VPC.country_id = VC.country_id;

select * from v_movies_countries_lf
```

#### 3. Right

muestra la tabla competa de la derecha (Right) sin considerar si hay o no valores que concatenan en la primera tabla, por ese motivo se ven "null" en las culumnas que pertenece a movies.


```sql
create or replace temp view v_movies_countries_rg
as
select VM.title,VM.budget,VM.revenue,VM.year_release_date,VM.release_date,VC.country_name
from v_movies_2011 VM
right join v_productions_countries VPC on VM.movie_id = VPC.movie_id
right join v_countries VC on VPC.country_id = VC.country_id;

select * from v_movies_countries_rg
```

#### Full 
Esta funcion concatena ls tablas de ambas tablas sin considerar si hay o no relación entre las tablas, indiscriminadamenete.


```sql
create or replace temp view v_movies_countries_fl
as
select VM.title,VM.budget,VM.revenue,VM.year_release_date,VM.release_date,VC.country_name
from v_movies_2011 VM
full join v_productions_countries VPC on VM.movie_id = VPC.movie_id
full join v_countries VC on VPC.country_id = VC.country_id;

select * from v_movies_countries_fl
```

#### Semi

Semi join unicamente nos denuelve valores que concatenan entre las 2 tablas pero solo muestra la primera tala

```sql
create or replace temp view v_movies_countries_sm
as
select VM.title,VM.budget,VM.revenue,VM.year_release_date,VM.release_date
from v_movies_2011 VM
inner join v_productions_countries VPC on VM.movie_id = VPC.movie_id
semi join v_countries VC on VPC.country_id = VC.country_id;

select * from v_movies_countries_sm
```

#### Anti
nos muestra los valores en donde no se han establecido el Join, de igual manera que el Semi Join no mostraremos la tabla 2 

```sql
create or replace temp view v_movies_countries_at
as
select VM.title,VM.budget,VM.revenue,VM.year_release_date,VM.release_date
from v_movies_2011 VM
inner join v_productions_countries VPC on VM.movie_id = VPC.movie_id
anti join v_countries VC on VPC.country_id = VC.country_id;

select * from v_movies_countries_at
```
