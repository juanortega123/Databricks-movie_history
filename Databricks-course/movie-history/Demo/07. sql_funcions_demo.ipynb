## Aggregate Funcions (Funciones de Agregación)

```sql
select year_release_date, count(1) as total_count, sum(budget) as sum_budget, max(budget) max_budget, min(budget) min_budget, avg(budget) avg_budget
from movies
group by year_release_date
having count(1) >220
order by year_release_date;
```

## Windows functions (Funciones de ventana)

```sql
select
  title,
  year_release_date, 
  release_date,
  rank() over(partition by year_release_date order by release_date) as rank
from movies
where year_release_date is not null
order by year_release_date;
```

```sql
select 
  title,
  year_release_date,
  release_date,
  rank() over(partition by year_release_date order by release_date)
from movies
where year_release_date is not null
order by release_date;
```
