realizar un analisis sobre el presupuesto e ingreso por productora.

```sql
use movie_gold
```

```sql
select company_name,
      count(company_name) as total_for_company,
      sum(budget)as total_budget,
      cast(avg(budget)as decimal (18,2))as avg_budget,
      sum(revenue)as total_revenue,
      cast(avg(revenue) as decimal (18,2)) as avg_revenue
from results_movie
where year_release_date between 2010 and 2015
group by company_name
order by avg_revenue desc
limit (15);
```
